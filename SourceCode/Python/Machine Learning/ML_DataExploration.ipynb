```python
#Import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.rcParams['figure.dpi'] = 200
```

```python
#Load dataset
df = pd.read_csv('./data/online_shoppers_intention.csv')
df.head()
```

```python
#Data information
df.info()
```

```python
#Summary statistics
df.describe()
```

```python
#Check missing values
df.isnull().sum()
```

```python
#Check correlation using Heatmap
plt.subplots(figsize=(15,15))
sns.heatmap(df.corr(), annot = True, cmap= 'coolwarm')
```

```python
fig, axes = plt.subplots(1,2, figsize=(10, 8))

#Distribution of Revenue
ax1 = sns.countplot(x='Revenue', data = df, palette = 'hls',ax = axes[0])
ax1.set_title("Purchase or Not")

#Distribution of Weekend
ax2 = sns.countplot(x='Weekend', hue = 'Revenue', data = df, palette = 'hls', ax = axes[1])
ax2.set_title("Purchase on Weekend")
```

```python
fig, axes = plt.subplots(1,2, figsize=(18,7))

#VisitorType vs Revenue
ax1 = sns.countplot(x='VisitorType', hue='Revenue' , data = df, palette = 'hls', ax = axes[0])
ax1.set_title("VisitorType vs Revenue")

#Month vs Revenue
ax2 = sns.countplot(x='Month',hue='Revenue' , data = df, palette = 'hls', ax = axes[1], )
ax2.set_title("Month vs Revenue")
```

```python
#Split X and Y
X = df.drop('Revenue', axis=1)
y = df['Revenue']
```

```python
#Configs
font = {'size': 15}
plt.rc('font', **font)
plt.rc('axes', labelpad=20)
plt.rc('axes', titlepad=20)

#Draw boxplot to visualize outliers
plt.figure(figsize=(15, 14))
ax = sns.boxplot(data=X)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.title(u'Boxplot of the independent variables X')
plt.ylabel('Value of the normalized variable')
plt.xlabel('Variables Name')
plt.tight_layout(pad=2, w_pad=3.0)
```

```python
#Check distribution
fig, axes = plt.subplots(3,1, figsize=(16,13))

ax1 = sns.distplot(df["BounceRates"],color="green",ax = axes[0])
ax1.set_title("Distribution of Bounce Rates")
ax1.set_xlabel("")

ax2 = sns.distplot(df["PageValues"],color="blue",ax = axes[1])
ax2.set_title("Distribution of Page Values")
ax2.set_xlabel("")

ax3 = sns.distplot(df["ExitRates"],color="red",ax = axes[2])
ax3.set_title("Distribution of Exit Rates")
ax3.set_xlabel("")
```
